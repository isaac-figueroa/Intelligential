## Inicialización del objeto pago

## Visualización de datos

```python
def imprimir(registros, limite=30):
    df = pd.DataFrame(registros)
    if len(registros)>0:
        df.columns = registros[0].keys()
    return df.head(limite)
```

### Visualización de todos los registros

```python
registros = pago.obtenerRegistros()
imprimir(registros)
```

### Visualización de los registros activos de un contrato

```python
registros = pago.obtenerRegistros(2)
imprimir(registros)
```

### Visualización de todos los registros de un contrato

```python
registros = pago.obtenerRegistros(2, soloActivos=False)
imprimir(registros)
```
